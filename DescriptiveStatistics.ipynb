```python
import numpy as np
import pandas as pd
import plotly.express as px
from scipy.stats import skew, kurtosis
import plotly.figure_factory as ff
```

```python
path = r"C:\Users\ASUS\Desktop\Campus\3rd Year\second\Nature Inspired Algorithms\Dataset\heart.csv"
df = pd.read_csv(path)
# Display the first few rows of the DataFrame
df.head()
```

```python
df1 = df[['age', 'sex', 'chol', 'fbs']]
df1.head()
```

```python
df1.describe()
```

```python
age = df['age']
hist_data = [age]
group_labels = ['Age']
fig = ff.create_distplot(hist_data, group_labels)
fig.update_layout(title_text='Age Distribution', xaxis_title='Age', yaxis_title='Density')
fig.show()
```

```python
print(df1['age'].mean())
print(df1['age'].median())
print(df1['age'].mode())
print(skew(df1['age']))
print(kurtosis(df1['age']))
```

```python
# Outlier calculation
Q1 = df["chol"].quantile(0.25)
Q3 = df["chol"].quantile(0.75)
IQR = Q3 - Q1
outlier1 = (Q1 - 1.5 * IQR)
outlier2 = (Q3 + 1.5 * IQR)
print("Lower outlier threshold:", outlier1)
print("Upper outlier threshold:", outlier2)
```

```python
fig = px.box(df, y="chol", points="all")
fig.update_layout(title_text='Cholesterol Box Plot', yaxis_title='Cholesterol Level')
fig.show()
```
